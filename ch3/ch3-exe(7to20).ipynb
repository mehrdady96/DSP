3.7.
<div dir="rtl" style="text-align: right; font-family: Tahoma; font-size: 16px;">

### **ورودی یک سیستم LTI علّی**  
ورودی این سیستم به‌صورت زیر تعریف شده است:  
$$
x[n] = u[-n - 1] + \left(\frac{1}{2}\right)^n u[n].
$$  

### **تبدیل Z خروجی سیستم**  
تبدیل Z خروجی این سیستم برابر است با:  
$$
Y(z) = \frac{-\frac{1}{2}z^{-1}}{\left(1 - \frac{1}{2}z^{-1} \right)(1 + z^{-1})}.
$$  

### **(الف) محاسبه \( H(z) \)، تبدیل Z پاسخ ضربه‌ای سیستم**  
باید تابع تبدیل سیستم را به دست آوریم و ناحیه همگرایی (ROC) را مشخص کنیم.  

### **(ب) تعیین ناحیه همگرایی \( Y(z) \)**  
ناحیه همگرایی (ROC) این تابع را بیابید.  

### **(ج) تعیین \( y[n] \)**  
پاسخ سیستم در حوزه زمان را مشخص کنید.  

</div>


<div dir="rtl" style="text-align: right; font-family: Tahoma; font-size: 16px;">

## **(الف) محاسبه \( H(z) \)**
ورودی سیستم به‌صورت زیر داده شده است:
$$
x[n] = u[-n - 1] + \left(\frac{1}{2}\right)^n u[n]
$$
با گرفتن تبدیل Z:
$$
X(z) = \frac{-1}{1 - z^{-1}} + \frac{1}{1 - \frac{1}{2}z^{-1}}, \quad \frac{1}{2} < |z| < 1
$$
اکنون برای یافتن \( H(z) \)، از رابطه زیر استفاده می‌کنیم:
$$
H(z) = \frac{Y(z)}{X(z)}
$$
جایگذاری مقدار \( Y(z) \) و \( X(z) \):
$$
H(z) = \frac{-\frac{1}{2}z^{-1}}{\left(1 - \frac{1}{2}z^{-1} \right)(1 + z^{-1})} \cdot \frac{(1 - z^{-1})(1 - \frac{1}{2}z^{-1})}{-\frac{1}{2}z^{-1}}
$$
پس داریم:
$$
H(z) = \frac{1 - z^{-1}}{1 + z^{-1}}
$$
از آنجا که \( H(z) \) علّی است، ناحیه همگرایی (ROC) آن به‌صورت زیر است:
$$
|z| > 1
$$

---

## **(ب) تعیین ناحیه همگرایی \( Y(z) \)**  
یکی از قطب‌های \( X(z) \) که باعث شده بود ROC آن به کمتر از ۱ محدود شود، توسط صفر \( H(z) \) حذف شده است.  
بنابراین، ROC تابع \( Y(z) \) ناحیه‌ای در صفحه \( z \) است که دو شرط زیر را ارضا کند:
$$
|z| > \frac{1}{2} \quad \text{و} \quad |z| > 1
$$
پس نتیجه می‌گیریم که:
$$
Y(z) \quad \text{برای} \quad |z| > 1 \quad \text{همگرا است.}
$$

---

## **(ج) تعیین \( y[n] \)**  
با تجزیه \( Y(z) \) به کسرهای جزئی، داریم:
$$
Y(z) = \frac{-\frac{1}{3}}{1 - \frac{1}{2}z^{-1}} + \frac{\frac{1}{3}}{1 + z^{-1}}, \quad |z| > 1
$$
با گرفتن تبدیل Z معکوس، پاسخ زمانی \( y[n] \) برابر است با:
$$
y[n] = -\frac{1}{3} \left(\frac{1}{2}\right)^n u[n] + \frac{1}{3} (-1)^n u[n]
$$

### **تابع تبدیل یک سیستم LTI علّی**  
تابع سیستم برای یک **سیستم LTI علّی** به‌صورت زیر داده شده است:  
$$
H(z) = \frac{1 - z^{-1}}{1 + \frac{3}{4}z^{-1}}.
$$

<div dir="rtl" style="text-align: right; font-family: Tahoma; font-size: 16px;">

### **ورودی سیستم LTI**  
ورودی سیستم به‌صورت زیر تعریف شده است:  
$$
x[n] = \left(\frac{1}{3}\right)^n u[n] + u[-n - 1].
$$

### **سؤالات:**  
**(الف)** پاسخ ضربه‌ای سیستم \( h[n] \) را بیابید.  
**(ب)** خروجی \( y[n] \) را محاسبه کنید.  
**(ج)** آیا سیستم پایدار است؟ یعنی آیا \( h[n] \) مجموع‌پذیر مطلق است؟  

</div>



<div dir="rtl" style="text-align: right; font-family: Tahoma; font-size: 16px;">

### **تابع سیستم علی:**  
$$
H(z) = \frac{1 - z^{-1}}{1 + \frac{3}{4} z^{-1}}
$$

ورودی سیستم به‌صورت زیر تعریف شده است:  
$$
x[n] = \left(\frac{1}{3}\right)^n u[n] + u[-n - 1].
$$

پس تبدیل \( z \) ورودی به‌صورت زیر است:  
$$
X(z) = \frac{1}{1 - \frac{1}{3} z^{-1}} - \frac{1}{1 - z^{-1}} = \frac{-\frac{2}{3} z^{-1}}{(1 - \frac{1}{3} z^{-1})(1 - z^{-1})}, \quad \frac{1}{3} < |z| < 1
$$

### **(الف) پاسخ ضربه‌ای \( h[n] \)**
چون سیستم علی است، پس:  
$$
h[n] = \left(-\frac{3}{4}\right)^n u[n] - \left(-\frac{3}{4}\right)^{n-1} u[n - 1]
$$

### **(ب) خروجی \( y[n] \)**
تبدیل \( z \) خروجی:  
$$
Y(z) = X(z) H(z) = \frac{-\frac{2}{3} z^{-1}}{(1 - \frac{1}{3} z^{-1})(1 + \frac{3}{4} z^{-1})}
$$
با تجزیه به کسرهای جزئی:  
$$
Y(z) = \frac{-\frac{8}{13}}{1 - \frac{1}{3} z^{-1}} + \frac{\frac{8}{13}}{1 + \frac{3}{4} z^{-1}}, \quad \frac{3}{4} < |z|
$$

بنابراین خروجی برابر است با:  
$$
y[n] = -\frac{8}{13} \left(\frac{1}{3}\right)^n u[n] + \frac{8}{13} \left(-\frac{3}{4}\right)^n u[n]
$$

### **(ج) بررسی پایداری**  
برای اینکه \( h[n] \) علی باشد، باید **ناحیه همگرایی** \( H(z) \) شرط زیر را داشته باشد:  
$$
\frac{3}{4} < |z|
$$
که **دایره واحد را شامل می‌شود**. پس \( h[n] \) **مجموع‌پذیر مطلق است** و **سیستم پایدار است**.

</div>

3.9.
<div dir="rtl" style="text-align: right; font-family: Tahoma; font-size: 16px;">

### **تابع تبدیل سیستم:**
$$
H(z) = \frac{1 + z^{-1}}{(1 - \frac{1}{2}z^{-1})(1 + \frac{1}{4}z^{-1})}
$$

### **(الف) ناحیه همگرایی (ROC)**
چون سیستم **علی** است، باید ناحیه همگرایی شامل **بیرون‌ترین قطب** باشد.  
قطب‌های مخرج عبارت‌اند از:  
$$
z = \frac{1}{2}, \quad z = -\frac{1}{4}
$$
برای سیستم علی، ناحیه همگرایی این شرط را دارد:
$$
|z| > \frac{1}{2}
$$

### **(ب) پایداری سیستم**
یک سیستم **پایدار** است اگر ناحیه همگرایی (ROC) شامل **دایره واحد** باشد، یعنی:  
$$
|z| > \frac{1}{2}
$$
چون **دایره واحد (یعنی $|z| = 1$) در این بازه قرار دارد، سیستم پایدار است.**

### **(ج) تبدیل \( z \) ورودی \( X(z) \)**
خروجی سیستم برابر است با:  
$$
y[n] = -\frac{1}{3} \left(\frac{1}{4}\right)^n u[n] - \frac{4}{3} (2)^n u[-n - 1]
$$
چون **\( Y(z) = X(z) H(z) \)** داریم:
$$
X(z) = \frac{Y(z)}{H(z)}
$$
با توجه به تجزیه به کسرهای جزئی و ویژگی‌های تبدیل **\( z \)**، می‌توان **\( X(z) \)** را یافت.

### **(د) پاسخ ضربه‌ای \( h[n] \)**
برای یافتن **\( h[n] \)**، ابتدا تابع تبدیل **\( H(z) \)** را به کسرهای جزئی تجزیه می‌کنیم:
$$
H(z) = \frac{1 + z^{-1}}{(1 - \frac{1}{2}z^{-1})(1 + \frac{1}{4}z^{-1})}
$$
با تجزیه به کسرهای جزئی:
$$
H(z) = \frac{A}{1 - \frac{1}{2}z^{-1}} + \frac{B}{1 + \frac{1}{4}z^{-1}}
$$
پس پاسخ ضربه‌ای **\( h[n] \)** برابر است با:
$$
h[n] = A \left(\frac{1}{2}\right)^n u[n] + B \left(-\frac{1}{4}\right)^n u[n]
$$
که در آن **ضرایب \( A \) و \( B \) از طریق حل معادلات حاصل از کسرهای جزئی تعیین می‌شوند.**

</div>


<div dir="rtl" style="text-align: right; font-family: Tahoma; font-size: 16px;">

### **تابع تبدیل سیستم:**
$$
H(z) = \frac{1 + z^{-1}}{(1 - \frac{1}{2}z^{-1})(1 + \frac{1}{4}z^{-1})} = \frac{2}{(1 - \frac{1}{2}z^{-1})} - \frac{1}{(1 + \frac{1}{4}z^{-1})}
$$

### **(الف) ناحیه همگرایی (ROC)**
چون سیستم **علی** است، ناحیه همگرایی شامل **بیرون‌ترین قطب** خواهد بود.  
قطب‌های مخرج عبارت‌اند از:
$$
z = \frac{1}{2}, \quad z = -\frac{1}{4}
$$
برای سیستم علی، ناحیه همگرایی شرط زیر را دارد:
$$
|z| > \frac{1}{2}
$$

### **(ب) پایداری سیستم**
سیستم زمانی **پایدار** است که ناحیه همگرایی شامل **دایره واحد** باشد.  
چون **دایره واحد (\(|z| = 1\)) در ناحیه همگرایی (\(|z| > \frac{1}{2}\)) قرار دارد، پس سیستم پایدار است. ✅

### **(ج) تبدیل \( z \) ورودی \( X(z) \)**
خروجی سیستم برابر است با:
$$
y[n] = -\frac{1}{3} \left(-\frac{1}{4}\right)^n u[n] - \frac{4}{3} (2)^n u[-n - 1]
$$
تبدیل **\( z \)** خروجی:
$$
Y(z) = -\frac{\frac{1}{3}}{1 + \frac{1}{4}z^{-1}} + \frac{\frac{4}{3}}{1 - 2z^{-1}}
$$
$$
Y(z) = \frac{1 + z^{-1}}{(1 + \frac{1}{4}z^{-1})(1 - 2z^{-1})}, \quad \frac{1}{4} < |z| < 2
$$
پس تبدیل **\( z \)** ورودی برابر است با:
$$
X(z) = \frac{Y(z)}{H(z)} = \frac{(1 - \frac{1}{2}z^{-1})}{(1 - 2z^{-1})}, \quad |z| < 2
$$
در نتیجه:
$$
x[n] = - (2)^n u[-n - 1] + \frac{1}{2} (2)^{n-1} u[-n]
$$

### **(د) پاسخ ضربه‌ای \( h[n] \)**
با تجزیه به کسرهای جزئی:
$$
H(z) = \frac{2}{(1 - \frac{1}{2}z^{-1})} - \frac{1}{(1 + \frac{1}{4}z^{-1})}
$$
پس:
$$
h[n] = 2 \left(\frac{1}{2}\right)^n u[n] - \left(-\frac{1}{4}\right)^n u[n]
$$

### **تحلیل ناحیه همگرایی (ROC) و تبدیل فوریه برای دنباله‌های داده شده**

#### **(الف)**
$$
x[n] = \left(\frac{1}{2}\right)^n + \left(\frac{3}{4}\right)^n u[n - 10]
$$
- اولین جمله **علی** است → \( |z| > \frac{1}{2} \)  
- دومین جمله نیز **علی** است اما از \( n = 10 \) فعال می‌شود → \( |z| > \frac{3}{4} \)  
- **ROC کلی:** \( |z| > \frac{3}{4} \)  
- **تبدیل فوریه همگرا است** زیرا ROC شامل \( |z| = 1 \) می‌شود.

#### **(ب)**
$$
x[n] =
\begin{cases}
1, & -10 \leq n \leq 10 \\
0, & \text{otherwise}
\end{cases}
$$
- دنباله **محدود** است → **کل صفحه \( z \) ناحیه همگرایی است، به جز \( z = 0 \) و \( z = \infty \)**  
- **تبدیل فوریه همگرا است.**

#### **(ج)**
$$
x[n] = 2^n u[-n]
$$
- دنباله **ضد علی** است → \( |z| < 2 \)  
- **تبدیل فوریه همگرا نیست** زیرا ROC شامل \( |z| = 1 \) نمی‌شود.

#### **(د)**
$$
x[n] = \left(\frac{1}{4}\right)^{n+4} - (e^{j\pi/3})^n \quad u[n - 1]
$$
- جمله اول یک نمایی کاهشی **علی** است → \( |z| > \frac{1}{4} \)  
- جمله دوم **علی** است → \( |z| > |e^{j\pi/3}| = 1 \)  
- **ROC کلی:** \( |z| > 1 \)  
- **تبدیل فوریه همگرا نیست.**

#### **(ه)**
$$
x[n] = u[n + 10] - u[n + 5]
$$
- دنباله **محدود** است → **کل صفحه \( z \) ناحیه همگرایی است، به جز \( z = 0 \) و \( z = \infty \)**  
- **تبدیل فوریه همگرا است.**

#### **(و)**
$$
x[n] = \left(\frac{1}{2}\right)^{n-1} u[n] + (2 + 3j)^{n-2} u[-n - 1]
$$
- جمله اول **علی** است → \( |z| > \frac{1}{2} \)  
- جمله دوم **ضد علی** است → \( |z| < |2 + 3j| = \sqrt{13} \)  
- **ROC کلی:** \( \frac{1}{2} < |z| < \sqrt{13} \)  
- **تبدیل فوریه همگرا است** زیرا \( |z| = 1 \) در ROC قرار دارد.

### **تحلیل سیستم‌های علی (Causal Systems)**  
اگر \( x[n] \) یک سیستم علی باشد، تبدیل z آن \( X(z) \) به صورت زیر تعریف می‌شود:  
$$
X(z) = \sum_{n=0}^{\infty} x[n]z^{-n}
$$  
این بدان معناست که این مجموع‌گیری شامل هیچ توان مثبتی از \( z \) نخواهد بود. بنابراین، فرم بسته‌ی \( X(z) \) باید در \( z = \infty \) همگرا باشد، یعنی \( z = \infty \) باید در ناحیه همگرایی (ROC) \( X(z) \) قرار داشته باشد، یا به عبارت دیگر:  
$$
\lim_{z \to \infty} X(z) \neq \infty
$$


**(الف)**  
$$
\lim_{z \to \infty} \frac{(1 - z^{-1})^2}{(1 - \frac{1}{2}z^{-1})} = 1 \quad \text{می‌تواند علی باشد}
$$

**(ب)**  
$$
\lim_{z \to \infty} \frac{(z - 1)^2}{(z - \frac{1}{2})} = \infty \quad \text{نمی‌تواند علی باشد}
$$

**(ج)**  
$$

### **رسم نمودار قطب-صفر و تعیین ناحیه همگرایی (ROC)**  

در این بخش، تبدیل‌های **Z** داده‌شده را بررسی کرده و نمودار قطب-صفر آن‌ها را ترسیم می‌کنیم. همچنین ناحیه همگرایی (ROC) را مشخص می‌کنیم.

---

**(الف)**  
تبدیل **Z** داده‌شده:  
$$
X_1(z) = \frac{1 - \frac{1}{2}z^{-1}}{1 + 2z^{-1}}
$$  
- **قطب‌ها:** \( z = -2 \)  
- **صفرها:** \( z = \frac{1}{2} \)  
- **ناحیه همگرایی:**  
  $$
  |z| < 2
  $$  

---

**(ب)**  
تبدیل **Z** داده‌شده:  
$$
X_2(z) = \frac{1 - \frac{1}{3}z^{-1}}{\left(1 + \frac{1}{2}z^{-1} \right) \left(1 - \frac{2}{3}z^{-1} \right)}
$$  
- **قطب‌ها:** \( z = -\frac{1}{2}, \quad z = \frac{2}{3} \)  
- **صفرها:** \( z = \frac{1}{3} \)  
- **چون سیستم علی است، ناحیه همگرایی از قطب خارجی گسترش می‌یابد:**  
  $$
  |z| > \frac{2}{3}
  $$  

---

**(ج)**  
تبدیل **Z** داده‌شده:  
$$
X_3(z) = \frac{1 + z^{-1} - 2z^{-2}}{1 - \frac{13}{6}z^{-1} + z^{-2}}
$$  
- **قطب‌ها:** \( z = \frac{3}{2}, \quad z = \frac{2}{3} \)  
- **صفرها:** \( z = 1, \quad z = -2 \)  
- **چون \( x_3[n] \) مجموع‌پذیر مطلق است، ناحیه همگرایی شامل دایره واحد است:**  
  $$
  \frac{2}{3} < |z| < \frac{3}{2}
  $$  

</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# تابع برای رسم نمودار قطب-صفر
def plot_pz(poles, zeros, roc, title):
    fig, ax = plt.subplots(figsize=(5, 5))

    # رسم دایره واحد
    unit_circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='dashed')
    ax.add_patch(unit_circle)

    # رسم قطب‌ها (با X قرمز)
    plt.scatter(np.real(poles), np.imag(poles), color='red', marker='x', s=100, label="Poles")

    # رسم صفرها (با O آبی)
    plt.scatter(np.real(zeros), np.imag(zeros), color='blue', marker='o', s=100, label="Zeros")

    # رسم ناحیه همگرایی (ROC)
    if roc == "inside":
        roc_circle = plt.Circle((0, 0), max(np.abs(poles)) * 1.1, color='gray', alpha=0.3)
    elif roc == "outside":
        roc_circle = plt.Circle((0, 0), min(np.abs(poles)) * 0.9, color='gray', alpha=0.3)
    elif roc == "between":
        inner_radius = min(np.abs(poles)) * 0.9
        outer_radius = max(np.abs(poles)) * 1.1
        theta = np.linspace(0, 2*np.pi, 100)
        x_inner = inner_radius * np.cos(theta)
        y_inner = inner_radius * np.sin(theta)
        x_outer = outer_radius * np.cos(theta)
        y_outer = outer_radius * np.sin(theta)
        ax.fill_betweenx(y_inner, x_inner, x_outer, color='gray', alpha=0.3)
    ax.add_patch(roc_circle)

    # تنظیمات نمودار
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.xlabel("Real")
    plt.ylabel("Imaginary")
    plt.title(title)
    plt.legend()
    plt.show()

# (الف) X1(z)
plot_pz(poles=[-2], zeros=[1/2], roc="inside", title="Pole-Zero Plot for X1(z)")

# (ب) X2(z)
plot_pz(poles=[-1/2, 2/3], zeros=[1/3], roc="outside", title="Pole-Zero Plot for X2(z)")

# (ج) X3(z)
plot_pz(poles=[3/2, 2/3], zeros=[1, -2], roc="between", title="Pole-Zero Plot for X3(z)")
```

$$  
$$
H(z) = \frac{1}{1 - \frac{1}{4}z^{-1}}, \quad |z| > \frac{1}{4}

**(ب)**

$$

### **مسئله 3.19**  
ناحیه همگرایی \( Y(z) \) شامل اشتراک ناحیه‌های همگرایی \( H(z) \) و \( X(z) \) است.

**(الف)**  
$$
Y(z) = \frac{1}{(1 + \frac{1}{2}z^{-1})(1 - \frac{1}{4}z^{-1})}
$$

اشتراک ناحیه‌های همگرایی \( H(z) \) و \( X(z) \) به صورت \( |z| > \frac{1}{2} \) است. بنابراین ناحیه همگرایی \( Y(z) \) برابر است با \( |z| > \frac{1}{2} \).

**(ب)**  
ناحیه همگرایی \( Y(z) \) دقیقاً برابر با اشتراک ناحیه‌های همگرایی \( H(z) \) و \( X(z) \) است: \( \frac{1}{3} < |z| < 2 \).

**(ج)**  
$$
Y(z) = \frac{1}{(1 - \frac{1}{5}z^{-1})(1 + \frac{1}{3}z^{-1})}
$$

ناحیه همگرایی \( Y(z) \) برابر است با \( |z| > \frac{1}{3} \).
