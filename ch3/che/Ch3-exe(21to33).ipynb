### **مسئله 3.27 - حل کامل**  
تبدیل z داده شده:

$$
X(z) = \frac{z^2}{(z - a)(z - b)} = \frac{z^2}{z^2 - (a + b)z + ab}
$$

**مرحله 1:** تبدیل به کسر مناسب:

$$
\frac{1}{z^2 - (a + b)z + ab} \left[ \begin{array}{ccc}
\frac{z^2}{z^2} & - & (a + b)z \\
(a + b)z & - & ab
\end{array} \right]
$$

**مرحله 2:** تجزیه به صورت زیر:

$$
X(z) = 1 + \frac{(a + b)z - ab}{(z - a)(z - b)} = 1 + \frac{\frac{(a + b)a - ab}{a - b}}{z - a} + \frac{\frac{(a + b)b - ab}{b - a}}{z - b}
$$

**مرحله 3:** ساده‌سازی عبارت:

$$
= 1 + \frac{\frac{a^2}{a - b}}{z - a} - \frac{\frac{b^2}{a - b}}{z - b} = 1 + \frac{1}{a - b} \left( \frac{a^2 z^{-1}}{1 - az^{-1}} - \frac{b^2 z^{-1}}{1 - bz^{-1}} \right)
$$
