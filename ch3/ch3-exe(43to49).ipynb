```python
# %% [python]
import matplotlib.pyplot as plt
import numpy as np

# ایجاد نمودار قطب-صفر
def plot_pole_zero(poles, zeros, title):
    plt.figure(figsize=(6,6))

    # رسم دایره واحد
    unit_circle = plt.Circle((0,0), 1, fill=False, color='black', linestyle='dashed')
    plt.gca().add_patch(unit_circle)

    # رسم قطب‌ها (×) و صفرها (○)
    plt.scatter(np.real(poles), np.imag(poles), marker='x', s=100, color='r')
    plt.scatter(np.real(zeros), np.imag(zeros), marker='o', s=100, color='b', facecolors='none')

    # تنظیمات نمودار
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.title(title)
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.axis('equal')
    plt.xlim((-1.5,1.5))
    plt.ylim((-1.5,1.5))
    plt.show()

# مثال برای شکل P3.48-1 (مقادیر نمونه)
plot_pole_zero(poles=[0.5+0.5j, 0.5-0.5j], zeros=[-0.7], title='Figure P3.48-1')

# مثال برای شکل P3.48-2 (مقادیر نمونه)
plot_pole_zero(poles=[0.8], zeros=[0.3+0.8j, 0.3-0.8j], title='Figure P3.48-2')
```

\mathcal{Y}(z) = -\frac{\sum_{k=1}^{N} a_k \left( \sum_{m=1}^{k} y[m - k - 1]z^{-m+1} \right)}{\sum_{k=0}^{N} a_k z^{-k}} + \frac{\sum_{k=0}^{M} b_k z^{-k}}{\sum_{k=0}^{N} a_k z^{-k}} \mathcal{X}(z)
$$

#### (ب) تجزیه خروجی:
$$
y[n] = y_{\text{ZIR}}[n] + y_{\text{ZICR}}[n]
$$

-$$ \( y_{\text{ZIR}}[n] \)$$: پاسخ به شرایط اولیه (ورودی صفر)

-$$ \( y_{\text{ZICR}}[n] \)$$: پاسخ به ورودی (شرایط اولیه صفر)

#### (ج) حالت شرایط اولیه صفر:
وقتی شرایط اولیه صفر باشند:
$$
\mathcal{Y}(z) = \frac{\sum_{k=0}^{M} b_k z^{-k}}{\sum_{k=0}^{N} a_k z^{-k}} \mathcal{X}(z)
$$
که با نتیجه تبدیل z دوطرفه یکسان است.

### مسئله ۳.۴۹ - حل کامل

#### قسمت (الف): تبدیل z یکطرفه خروجی
برای سیستم با شرایط اولیه غیرصفر:
$$
\mathcal{Y}(z) = -\frac{\sum_{k=1}^{N} a_k \left( \sum_{m=1}^{k} y[m-k-1]z^{-m+1} \right)}{A(z)} + \frac{B(z)}{A(z)}\mathcal{X}(z)
$$
که در آن:
$$
A(z) = \sum_{k=0}^{N} a_k z^{-k}, \quad B(z) = \sum_{k=0}^{M} b_k z^{-k}
$$

#### قسمت (ب): تجزیه پاسخ سیستم
پاسخ کلی سیستم به صورت:
$$
y[n] = y_{\text{ZIR}}[n] + y_{\text{ZICR}}[n]
$$

- $$y_{\text{ZICR}}[n]$$: پاسخ به ورودی (شرایط اولیه=۰)

$$
\mathcal{Y}(z) = H(z)\mathcal{X}(z) = \frac{B(z)}{A(z)}\mathcal{X}(z)
$$
که با نتیجه تبدیل z دوطرفه یکسان است.
